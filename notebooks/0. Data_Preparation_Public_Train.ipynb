```python
import cv2

def resize_img(path):
  try:
    img = cv2.imread(path)
#     img = p(img)
    img = cv2.resize(img, (224, 224))
    img = img.astype(np.float32)/255
    return img
  except Exception as e:
    print(str(e))
    return None
```
